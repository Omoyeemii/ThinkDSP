```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Load audio
audio_path = "c.wav"
y, sr = librosa.load(audio_path, sr=None)

# If stereo, take left channel
if len(y.shape) > 1:
    y = y[:, 0]

# Time axis
time = np.linspace(0, len(y)/sr, len(y))

# Set up plot
fig, ax = plt.subplots(figsize=(12, 4))
ax.set_ylim(-1, 1)
ax.set_xlim(0, 5)  # 5-second window
ax.set_title("Waveform Animation")
ax.set_xlabel("Time (s)")
ax.set_ylabel("Amplitude")

line, = ax.plot([], [])

# Animation function
window_size = sr * 5  # 5 seconds of samples per frame
def update(frame):
    start = frame * window_size
    end = start + window_size
    if end > len(y):
        end = len(y)
    line.set_data(time[start:end], y[start:end])
    return line,

frames = len(y) // window_size
ani = FuncAnimation(fig, update, frames=frames, interval=200, blit=True)

# Save as GIF
ani.save("waveform_animation.gif", writer='pillow')
print("Waveform animation saved as GIF!")
```
